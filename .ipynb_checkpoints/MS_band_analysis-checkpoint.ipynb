```python
patients = []
plist = os.listdir('./MS_Band_Raw/')

for m in range(1,len(plist)):
    list = os.listdir('./MS_Band_Raw/%d_logger_data' % m)
    if m == 7:
        continue
    print(m, 'Participant has been read')
    for i in range(1, len(list)):
        p = pd.read_csv('./MS_Band_Raw/%d_logger_data/Day%d.csv' % (m,i), skiprows=1)
        p.insert(0, 'PID', m, allow_duplicates=False)
        p.insert(1,'DAY',i, allow_duplicates=False)
        if 8< len(p['Time Stamp'].loc[1]) < 11:
            p['Time Stamp'] = pd.to_datetime(p['Time Stamp'], format='%H:%M:%S %p').dt.time
        elif len(p['Time Stamp'].loc[1]) > 11:
            p['Time Stamp'] = pd.to_datetime(p['Time Stamp'], format='%H:%M:%S.%f').dt.time
        else:
            p['Time Stamp'] = pd.to_datetime(p['Time Stamp']).dt.time
#        p['Time Stamp'] = pd.to_timestamp(p['Time Stamp'])
        patients.append(p)
        print(i, 'Day has been read')
```

```python
patients.insert(3,'TD')
for t in patients['Time Stamp']:
    patients['TD'].loc[t] = patients['Time Stamp'].loc[t+1] - patients['Time Stamp'].loc[t]
```
